This notebook demonstrates how to use the Azure OpenAI service to generate tokens and embeddings for a given text.
```python
# load openai
from openai import AzureOpenAI
import tiktoken

client = AzureOpenAI(
  api_key="e5d08a92522448088b778c4d0dc4e2e0",
  api_version = "2024-02-15-preview",
  azure_endpoint="https://ai-svc-swe394158187034.openai.azure.com/"
)
```

```python
# use tiktoken to generate tokens

#tokenizer = tiktoken.get_encoding("gpt-4")
tokenizer = tiktoken.get_encoding("o200k_base")

# use the tokenizer to encode text
text = "Once upon a time,"

encoding = tokenizer.encode(text)
print(encoding)

# use the tokenizer to decode text
decode = tokenizer.decode_tokens_bytes(encoding)
print(decode)

```

```python
# create embeddings from the tokens

def get_embedding(text):
    return client.embeddings.create(input = [text], model="text-embedding-ada-002").data[0].embedding

embeddings = get_embedding(encoding)
print(embeddings)


# embeddings = client.embeddings.create(
#     model="text-embedding-ada-002",
#     input=encoding
#     )

# print(embeddings)

# print(embeddings.model_dump_json(indent=2))
```

```python

embeddings_text = " "
for embedding in embeddings:
    embeddings_text += str(embedding) + " "

response = client.completions.create(
    model="text-embedding-ada-002",
    prompt=f"Generate a chat response based on these embeddings: {embeddings_text}",
    max_tokens=800,
    stop=None
)
print(response.choices[0].text.strip())


#print(resp)
#completion = response.choices[0].text.strip()
#print(completion)





```
